```python
#11.Create a dataframe by performing a group by over age_group and calculate the count of total samples under each age_group.
data.head(2)
```

```python
data_age = pd.DataFrame(data.groupby('age_group').response_id.count())
data_age.reset_index(inplace=True)
data_age.rename(columns={'response_id':'total_num'}, inplace=True)
data_age
```

```python
#12.Create a dataframe by performing a group by over age_group and finding the count of total samples for each age_group that opted for the JD(U) party in Vote_Now.
data.head(2)
```

```python
data7 = pd.DataFrame(data[data['Vote_Now']=='JD(U)'].groupby('age_group').response_id.count())
data7.reset_index(inplace=True)
data7.rename(columns={'response_id':'JD(U)_num'}, inplace=True)
data7
```

```python
#13.Join/Merge the two dataframes from questions 12 and 11 with the common column as age_group.
```

```python
final_set = pd.merge(left=data7, right=data_age, on='age_group')
final_set
```
