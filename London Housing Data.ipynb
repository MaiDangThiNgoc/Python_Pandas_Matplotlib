#5. What is the maximum & minimum 'average_price' per year in England ?

```python
data.head(2)
```

```python
data_england = data.groupby('area').get_group('england')
```

```python
data_england['year']=data_england['date'].dt.year
```

```python
data_england.groupby('year').average_price.agg(['min', 'max'])
```
